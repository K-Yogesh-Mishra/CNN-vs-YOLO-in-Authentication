```python
import os
import cv2

# Function to convert YOLO format to required format
def yolo_to_standard(yolo_annotation, img_width, img_height):
    class_id, x_center, y_center, width, height = map(float, yolo_annotation.split())
    left_x = int((x_center - width / 2) * img_width)
    top_y = int((y_center - height / 2) * img_height)
    width = int(width * img_width)
    height = int(height * img_height)
    return class_id, left_x, top_y, width, height

# Function to load class names from a file
def load_class_names(class_file):
    with open(class_file, 'r') as f:
        classes = f.read().strip().split('\n')
    return classes

# Directory paths
annotations_dir = '/content/gdrive/My Drive/YoloV8_2/data/images/train'  # YOLO format annotations
images_dir = '/content/gdrive/My Drive/YoloV8_2/data/images/train'  # Corresponding images
class_file = '/content/gdrive/My Drive/YoloV8_2/data/images/train/classes.txt'  # File containing class names
test_file = '/content/gdrive/My Drive/YoloV8_2/data/images/train/test.txt'  # File containing test image names
output_file = '/content/gdrive/My Drive/YoloV8_2/data/images/train/ground_truth.txt'

# Load class names
class_names = load_class_names(class_file)

# Load test image names
with open(test_file, 'r') as f:
    test_images = f.read().strip().split('\n')

# Process each image listed in test.txt
with open(output_file, 'w') as output:
    for image_name in test_images:
        annotation_file = os.path.splitext(image_name)[0] + '.txt'  # Adjust based on your annotation format
        image_path = os.path.join(images_dir, image_name)
        annotation_path = os.path.join(annotations_dir, annotation_file)

        # Get image dimensions
        img = cv2.imread(image_path)
        if img is None:
            print(f"Error: Unable to read image file {image_path}")
            continue  # Skip this image and annotation if the image can't be loaded

        img_height, img_width = img.shape[:2]

        with open(annotation_path, 'r') as f:
            annotations = f.readlines()

        output.write(f'{image_path}\n')
        for annotation in annotations:
            class_id, left_x, top_y, width, height = yolo_to_standard(annotation, img_width, img_height)
            class_name = class_names[int(class_id)]
            output.write(f'{class_name} {left_x} {top_y} {width} {height}\n')
```

```python
import re

# Path to the result file generated by YOLO
result_file = '/content/gdrive/My Drive/YoloV8_2/data/images/train/result.txt'
formatted_result_file = '/content/gdrive/My Drive/YoloV8_2/data/images/train/formatted_result.txt'

# Regular expression patterns to extract information
image_path_pattern = re.compile(r'(/content/.+\.jpg): Predicted')
prediction_pattern = re.compile(r'(\w+): \d+%\s+\(left_x:\s*(\d+)\s+top_y:\s*(\d+)\s+width:\s*(\d+)\s+height:\s*(\d+)\)')

# Read the result file
with open(result_file, 'r') as file:
    lines = file.readlines()

# Process the lines to extract and format the predictions
with open(formatted_result_file, 'w') as output_file:
    for line in lines:
        image_match = image_path_pattern.search(line)
        if image_match:
            image_path = image_match.group(1)
            output_file.write(f'{image_path}\n')

        prediction_match = prediction_pattern.search(line)
        if prediction_match:
            class_name, left_x, top_y, width, height = prediction_match.groups()
            output_file.write(f'{class_name} {left_x} {top_y} {width} {height}\n')

print(f"Formatted results saved to {formatted_result_file}")
```

```python
import os

def parse_annotations(file_path):
    annotations = {}
    with open(file_path, 'r') as file:
        lines = file.readlines()
        current_image = None
        for line in lines:
            if line.startswith('/'):
                current_image = line.strip()
                annotations[current_image] = []
            else:
                parts = line.strip().split()
                class_name = parts[0]
                bbox = list(map(int, parts[1:]))
                annotations[current_image].append((class_name, bbox))
    return annotations

ground_truth_file = '/content/gdrive/My Drive/YoloV8_2/data/images/train/ground_truth.txt'
formatted_result_file = '/content/gdrive/My Drive/YoloV8_2/data/images/train/formatted_result.txt'

ground_truth = parse_annotations(ground_truth_file)
formatted_result = parse_annotations(formatted_result_file)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix, classification_report,accuracy_score
import matplotlib.pyplot as plt
import numpy as np

def extract_labels_and_predictions(ground_truth, formatted_result):
    y_true = []
    y_pred = []

    for image_path, gt_annotations in ground_truth.items():
        gt_labels = [ann[0] for ann in gt_annotations]
        if image_path in formatted_result:
            pred_annotations = formatted_result[image_path]
            pred_labels = [ann[0] for ann in pred_annotations]
        else:
            pred_labels = []

        # Assuming one-to-one correspondence in number of boxes for simplicity
        # Otherwise, need a more sophisticated matching based on IoU, etc.
        max_len = max(len(gt_labels), len(pred_labels))
        gt_labels.extend(['none'] * (max_len - len(gt_labels)))
        pred_labels.extend(['none'] * (max_len - len(pred_labels)))

        y_true.extend(gt_labels)
        y_pred.extend(pred_labels)

    return y_true, y_pred

y_true, y_pred = extract_labels_and_predictions(ground_truth, formatted_result)

# Calculate metrics
precision = precision_score(y_true, y_pred, average='weighted', zero_division=0)
recall = recall_score(y_true, y_pred, average='weighted', zero_division=0)
f1 = f1_score(y_true, y_pred, average='weighted', zero_division=0)
accuracy = accuracy_score(y_true, y_pred)
conf_matrix = confusion_matrix(y_true, y_pred, labels=['doctor', 'nurse', 'patient', 'none'])

print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1-Score: {f1}')
print(f'Accuracy: {accuracy}')
print(f'Confusion Matrix:\n{conf_matrix}')
print('\nClassification Report:\n', classification_report(y_true, y_pred, labels=['doctor', 'nurse', 'patient', 'none']))
metrics = {
    'Precision': precision,
    'Recall': recall,
    'F1-Score': f1
}

# Extracting values for each class
report = classification_report(y_true, y_pred, labels=['doctor', 'nurse', 'patient', 'none'], output_dict=True)
classes = list(report.keys())[:-3]  # Exclude 'accuracy', 'macro avg', 'weighted avg'
class_metrics = {cls: [] for cls in classes}

for cls in classes:
    class_metrics[cls].append(report[cls]['precision'])
    class_metrics[cls].append(report[cls]['recall'])
    class_metrics[cls].append(report[cls]['f1-score'])

# Plot overall metrics
plt.figure(figsize=(10, 6))
plt.bar(metrics.keys(), metrics.values(), color=['blue', 'orange', 'green'])
plt.xlabel('Metrics')
plt.ylabel('Scores')
plt.title('Overall Metrics')
plt.ylim([0, 1])
plt.show()

# Plot metrics for each class
x_labels = ['Precision', 'Recall', 'F1-Score']
x = np.arange(len(x_labels))
width = 0.2

fig, ax = plt.subplots(figsize=(12, 8))
for i, cls in enumerate(classes):
    ax.bar(x + i*width, class_metrics[cls], width, label=cls)

ax.set_xlabel('Metrics')
ax.set_ylabel('Scores')
ax.set_title('Metrics by Class')
ax.set_xticks(x + width / 2)
ax.set_xticklabels(x_labels)
ax.legend()
plt.ylim([0, 1])
plt.show()
```

```python
import cv2
import networkx as nx
import matplotlib.pyplot as plt

# Load the model
net = cv2.dnn.readNetFromDarknet("/content/gdrive/My Drive/YoloV8_2/darknet/cfg/yolov4_train.cfg")

# Get the layer names
layer_names = net.getLayerNames()

# Create a directed graph
G = nx.DiGraph()

# Add a node for each layer
for i, layer_name in enumerate(layer_names):
    G.add_node(i, name=layer_name)

# Add edges between layers (assuming each layer is connected to the next)
for i in range(len(layer_names) - 1):
    G.add_edge(i, i + 1)

# Get a list of degrees for all nodes
degrees = [G.degree(n) for n in G.nodes()]

# Draw the graph
plt.figure(figsize=(20, 20))  # Increase figure size
pos = nx.spring_layout(G, seed=42)  # positions for all nodes

# Use different colors for nodes and varying node sizes based on their degree
nx.draw(G, pos, with_labels=True, font_weight='bold', node_color=degrees, node_size=[n * 100 for n in degrees], font_size=8)

# Show the plot
plt.show()
```
